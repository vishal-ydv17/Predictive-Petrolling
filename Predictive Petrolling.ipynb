```python
#T1 represents the day , month , year
#T2 represents the hr , min , sec
LDF[['T1' , 'T2']] = LDF.OCCURRED_ON_DATE.str.split(expand=True)
```

```python
LDF[['Year' , 'Month' , 'DateOfMonth']] = LDF.T1.str.split("-" , expand=True)
LDF[['Hr' , 'Min' , 'Sec']] = LDF.T2.str.split(":" , expand=True)
```

```python
LDF.drop(['OCCURRED_ON_DATE' , 'YEAR' , 'MONTH' , 'HOUR' , 'STREET' , 'T1' , 'T2'] , axis = 1 , inplace=True)
LDF
```

```python
LDF['OFFENSE_CODE_GROUP'].unique()
```

```python
LDF['OFFENSE_CODE_GROUP']=LDF['OFFENSE_CODE_GROUP'].astype('category')
LDF['OFFENSE_CODE_GROUP']=LDF['OFFENSE_CODE_GROUP'].cat.codes
LDF.head(15)
```

```python
x = LDF[['OFFENSE_CODE_GROUP' , 'DAY_OF_WEEK' , 'Lat' , 'Long' , 'Month' , 'DateOfMonth' , 'Hr' , 'Min' , 'Sec']]
y = LDF[['Location']]
```
